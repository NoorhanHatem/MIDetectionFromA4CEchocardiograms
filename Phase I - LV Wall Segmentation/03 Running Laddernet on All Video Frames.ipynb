```python
import os
import numpy as np
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
def process_and_predict(model, frame_dir, output_base_dir):
    # Get list of sub-directories (each represents a distinct video)
    video_dirs = [d for d in os.listdir(frame_dir) if os.path.isdir(os.path.join(frame_dir, d))]
    
    for video_dir in video_dirs:
        video_path = os.path.join(frame_dir, video_dir)
        output_dir = os.path.join(output_base_dir, video_dir)
        os.makedirs(output_dir, exist_ok=True)  
        
        # Process each frame in the sub-directory
        for frame_name in os.listdir(video_path):
            frame_path = os.path.join(video_path, frame_name)
            if frame_path.endswith(('.png', '.jpg', '.jpeg')):
                frame = cv2.imread(frame_path)
                frame = cv2.resize(frame, (224, 224))  
                frame = frame / 255.0  # Normalize frame to range 0-1
                frame = np.expand_dims(frame, axis=0)  
                
                pred = model.predict(frame)
                pred_mask = np.argmax(pred, axis=-1)[0]  # Remove batch dimension
                
                output_path = os.path.join(output_dir, frame_name)
                plt.imsave(output_path, pred_mask, cmap='gray')

model_path = r"C:\Users\noorh\Echocardiograms\Final Submission\laddernet_model.keras"
frame_dir = r"C:\Users\noorh\OneDrive\Desktop\Grad Project\HMC Dataset\Extracted Frames"
output_base_dir = r"C:\Users\noorh\OneDrive\Desktop\Grad Project\HMC Dataset\Predicted Segmentation Masks"
```

```python
model = tf.keras.models.load_model(model_path)

process_and_predict(model, frame_dir, output_base_dir)
```
